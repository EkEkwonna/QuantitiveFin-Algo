FUTURE VALUE OF MONEY 

********

1.DISCRETE

FV = PV (1+r)^n

Future value(FV) of a given amount is equal to Present Value(PV) of a given amount adding interest rate multiplied (by itself by) a number of years. 
Higher the value of the duration the higher the future value 

PV = FV / (1+r)^n

Present value (PV) is equal to the Future Value (FV) divided by (1+r )^n

2.CONTINUOUS 

Continuous model, we would have to deal with a differential equation 

x(t) = amount of money in the bank at time t
dt = difference in time 

How much money we made over dt (time step) is : x(t+dt) - x(t)
Difference between the money in the future (t + dt) and money at time t


Using Taylor's Expansion we can get the following: 

x(t+dt) - x(t) = (dx(t)/dt )dt

x(t+dt) - x(t) = r x(t) dt

The interest/premium I recieve is proportional to the actual amount of money multiplied by the interest rate. This means 

(dx(t)/dt) dt = rx(t) dt

The derivative of x (money in the bank) with respect to dt (difference in time) multiplied by the timedifference is equal to the interst rate
multiplied by the actual time multiplied by the timestamp which gives us

x(t) = x(0)e^(r*t)

fundamental principle is the same, (the longer we lend our money to a given entity for a longer time period we expect a higher return)

FV = PV * e^(-r*t)

STOCKS 

Stocks - Ownership of the small piece of a company 
Shares - are the units of stock 

You raise capital by selling off future profits in the form of shares. This is how small companies begin! 

First of all there are companies that want to raise capital for their business. Borrowing money from a bank can be too expensive so they may decide to go public instead. Allowing Investors to buy shares and stocks of underlying companies. 

Investors can't contact companies directly (they may not be interested or it's too much of a hassle) so the Stock Exchange came to be. 

Stock Exchange brings to gether buyers and sellers, it has a list of all the companies. 

Investors want to find the right company to invest in, so they use broker firms to have the ability to buy and sell shares. 


FLUCTUATION 

Fluctuations in stock prices are based on the fluctuation of Supply and Demand

If more investors want to buy a given stock then it's market price will increase, but if more investors want to sell a given stock it's market price will decrease. 


Why is it good to own a stock? 

There may be a growth in the value of the stock that can be realized if you sell the given stock. 
You also receive dividends - payments paid out to shareholders every quarter / 6 months (based on profitability of given company)
The more profitable a given company the higher the dividend 



VOLATILITY 

Measuring the risk of a stock 

* We can measure volatility with standard deviation (or variance) between returns of the same security (stock / bond etc)
* We can use the Capital Asset Pricing Model (CAPM) we can approximate the "risk associated with a given stock" with "beta"
