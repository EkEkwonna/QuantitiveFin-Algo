Time Value of Money

- the concept that money today is worht more than money in the future 

i.e. in a situation where we have an investor with a given amount of money an investor may choose to invest in : 
1. Treasury bills - (i.e. lend your money to the government)
2. Corporate bonds - lend your money to a company (corporate bonds or Stocks)

When you buy a bond you lend your money to a given entity in exchange for a given cashflow 

-------------

Money today is Worth more than the same amount of money in the future, because we may invest that moeny into government bonds or stocks, (future money is worht less because of inflation)

In the example you lend money to the Governemtn (Treasury Bills)

How much money you can make by lending your money? 
It depends on r (interest rate) and n (amount of time that will pass)

-------

FUTURE VALUE OF MONEY 

********

1.DISCRETE

FV = PV (1+r)^n

Future value(FV) of a given amount is equal to Present Value(PV) of a given amount adding interest rate multiplied (by itself by) a number of years. 
Higher the value of the duration the higher the future value 

PV = FV / (1+r)^n

Present value (PV) is equal to the Future Value (FV) divided by (1+r )^n

2.CONTINUOUS 

Continuous model, we would have to deal with a differential equation 

x(t) = amount of money in the bank at time t
dt = difference in time 

How much money we made over dt (time step) is : x(t+dt) - x(t)
Difference between the money in the future (t + dt) and money at time t


Using Taylor's Expansion we can get the following: 

x(t+dt) - x(t) = (dx(t)/dt )dt

x(t+dt) - x(t) = r x(t) dt

The interest/premium I recieve is proportional to the actual amount of money multiplied by the interest rate. This means 

(dx(t)/dt) dt = rx(t) dt

The derivative of x (money in the bank) with respect to dt (difference in time) multiplied by the timedifference is equal to the interst rate
multiplied by the actual time multiplied by the timestamp which gives us

x(t) = x(0)e^(r*t)

fundamental principle is the same, (the longer we lend our money to a given entity for a longer time period we expect a higher return)

FV = PV * e^(-r*t)
